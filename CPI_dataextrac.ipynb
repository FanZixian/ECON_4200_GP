```python
import pandas as pd

file_path = "./Data/CPI.txt"
with open(file_path, "r") as file:
    lines = file.readlines()

data = {
    "Date": [],
    "All items": [],
    "Food": [],
    "Food at home": [],
    "Food away from home": [],
    "Energy": [],
    "Gasoline (all types)": [],
    "Electricity": [],
    "Natural gas (piped)": [],
    "All items less food and energy": [],
    "Commodities less food and energy commodities": [],
    "Apparel": [],
    "New vehicles": [],
    "Medical care commodities": [],
    "Services less energy services": [],
    "Shelter": [],
    "Medical care services": [],
    "Education and communication": []
}

current_date = None
for line in lines:
    line = line.strip()
    if (line.startswith("Sept") or line.startswith("Oct") or line.startswith("Nov") 
            or line.startswith("Dec") or line.startswith("Jan") or line.startswith("Feb")
            or line.startswith("Mar") or line.startswith("Apr") or line.startswith("May") 
            or line.startswith("June") or line.startswith("July") or line.startswith("Aug")):
        current_date = line
    elif line and not line.startswith("Month"):
        values = line.split("\t")
        # print(line)
        data["Date"].append(current_date)
        data["All items"].append(float(values[0].strip('%')) / 100)
        data["Food"].append(float(values[1].strip('%')) / 100)
        data["Food at home"].append(float(values[2].strip('%')) / 100)
        data["Food away from home"].append(float(values[3].strip('%')) / 100)
        data["Energy"].append(float(values[4].strip('%')) / 100)
        data["Gasoline (all types)"].append(float(values[5].strip('%')) / 100)
        data["Electricity"].append(float(values[6].strip('%')) / 100)
        data["Natural gas (piped)"].append(float(values[7].strip('%')) / 100)
        data["All items less food and energy"].append(float(values[8].strip('%')) / 100)
        data["Commodities less food and energy commodities"].append(float(values[9].strip('%')) / 100)
        data["Apparel"].append(float(values[10].strip('%')) / 100)
        data["New vehicles"].append(float(values[11].strip('%')) / 100)
        data["Medical care commodities"].append(float(values[12].strip('%')) / 100)
        data["Services less energy services"].append(float(values[13].strip('%')) / 100)
        data["Shelter"].append(float(values[14].strip('%')) / 100)
        data["Medical care services"].append(float(values[15].strip('%')) / 100)
        data["Education and communication"].append(float(values[16].strip('%')) / 100)

df = pd.DataFrame(data)

print(df)
```

```python
df.to_csv('./Data/Raw Data/CPI_fromtxt.csv')
```
